```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV
df = pd.read_csv('sales_data.csv')

# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Data overview
print(df.head())
print(df.describe())
print(df.isnull().sum())

# Group by Product
sales_per_product = df.groupby('Product')['Sales'].sum().reset_index()

# Group by Category
sales_per_category = df.groupby('Category')['Sales'].sum().reset_index()

# Monthly sales
monthly_sales = df.groupby(df['Date'].dt.month)['Sales'].sum().reset_index()
monthly_sales.columns = ['Month', 'Sales']

# Plot charts
plt.figure(figsize=(8,5))
sns.barplot(x='Product', y='Sales', data=sales_per_product)
plt.title("Total Sales per Product")
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(x='Category', y='Sales', data=sales_per_category)
plt.title("Total Sales per Category")
plt.show()

plt.figure(figsize=(7,4))
sns.lineplot(x='Month', y='Sales', data=monthly_sales, marker='o')
plt.title("Monthly Sales Trend")
plt.show()
```
